# Data Analysis
In this part I will analyse the data and try to interpret it further.  
It's not supposed to change the dataframe but just visualize and explore

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load and show data
math_df = pd.read_csv('Math_Test_Results_Cleaned.csv')
math_df
```

The column I'm going to predict is the Mean Scale Score.  
I will look at the distribution first

```python
plt.hist(math_df["Mean Scale Score"], bins=20)
plt.title('Mean Scale Score distribution')
plt.xlabel('Score')
plt.ylabel('# Schools with score')
plt.show()
```

The Mean Scale Score is normally distributed but there could be some outliers
Lets find out with a boxplot

```python
plt.figure(figsize =(10, 5))

bp = plt.boxplot(math_df['Mean Scale Score'], patch_artist = True,  vert = False)

plt.setp(bp["boxes"], facecolor="blue")
plt.setp(bp["whiskers"], color="blue", linewidth = 2)
plt.setp(bp["caps"], color="red", linewidth = 3)
plt.setp(bp["medians"], color="red")
plt.setp(bp["fliers"], marker ="x")

plt.xlabel("Score")
plt.title("MSC boxplot")

plt.show()
```

As you can see, there are a lot of outliers but no extreme values.  
I'm going to check how many outliers there are.

```python
Q1 = math_df.quantile(0.25)
Q3 = math_df.quantile(0.75)
IQR = Q3 - Q1

((math_df < (Q1 - 1.5 * IQR)) |(math_df > (Q3 + 1.5 * IQR))).sum()
```

There are 635 outliers for the Mean Scale Score column.  
I can remove some of those but I'm not going to to that right now.  
This can be helpful later and I can still remove them when needed.

Next, is the MSC correlated with any of the remaining features?  
I will create a correlation matrix to find this out

```python
plt.figure(figsize=(10, 8), dpi=80)
matrix = math_df.corr().round(2)
sns.heatmap(matrix, annot=True)
plt.show()
```

As you can see, there is a very high correlation between the Mean Scale Score and Pct Level 4  
This can be drawn in a separate graph as well

```python
x, y = math_df['Mean Scale Score'], math_df['Pct Level 4']

# Least squares polynomial fit.
m, b = np.polyfit(x, y, 1)

fig, axs = plt.subplots(2, figsize=(8, 12))
fig.suptitle('Correlation between MSC and Pct Level 4')

plt.ylabel("Pct Level 4")
plt.xlabel("Mean Scale Score")
axs[0].scatter(x, y)

axs[1].scatter(x, y)
axs[1].plot(x, m*x+b, color="red")
```

There is a strong positive correlation between these features  
The Num Level 4 column also has a good correlation but this is expected because it is derived from the Pct Level 4 column

```python
x, y = math_df['Mean Scale Score'], math_df['Num Level 4']

# Least squares polynomial fit.
m, b = np.polyfit(x, y, 1)

fig, axs = plt.subplots(2, figsize=(8, 12))
fig.suptitle('Correlation between MSC and Num Level 4')

plt.ylabel("Num Level 4")
plt.xlabel("Mean Scale Score")
axs[0].scatter(x, y)

axs[1].scatter(x, y)
axs[1].plot(x, m*x+b, color="red")
```

There is not as much correlation but still something similar going on  
Now check the negative correlation with Pct Level 1

```python
x, y = math_df['Mean Scale Score'], math_df['Pct Level 1']

# Least squares polynomial fit.
m, b = np.polyfit(x, y, 1)

fig, axs = plt.subplots(2, figsize=(8, 12))
fig.suptitle('Correlation between MSC and Pct Level 1')

plt.ylabel("Pct Level 1")
plt.xlabel("Mean Scale Score")
axs[0].scatter(x, y)

axs[1].scatter(x, y)
axs[1].plot(x, m*x+b, color="red")
```

This is also highly negative correlated
